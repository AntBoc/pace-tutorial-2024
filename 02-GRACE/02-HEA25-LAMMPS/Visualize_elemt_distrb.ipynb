# HEA25S

```python
traj=read("10-elems-HEA25S-finished/dump.mdmc.dump", index=":")
```

```python
len(traj)
```

```python
at=traj[-1]
df=pd.DataFrame({"el":at.get_chemical_symbols(), "z": at.positions[:,2]})
gdf=df.groupby("el")["z"].agg(list)
```
